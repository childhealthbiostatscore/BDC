```python
# Libraries
import pandas as pd
import numpy as np
from redcap import Project, RedcapError
from tableone import TableOne
from datetime import datetime
# Data
tokens = pd.read_csv('/Users/timvigers/Documents/Work/api_tokens.csv')
token = tokens.iloc[1,1]
project = Project("https://redcap.ucdenver.edu/api/", token)
data = pd.DataFrame(project.export_records())
```

```python
# Get demographic info
demographics = data.drop_duplicates(subset='study_id', keep='first')
demographics = demographics.reset_index(drop=True)
# Combine race columns
race = []
for i in range(demographics.shape[0]):
    if demographics.loc[i,'race_multi'] == "1":
        race.append('race_multi')
    elif demographics.loc[i,'race___0'] == "1":
        race.append('race_0')
    elif demographics.loc[i,'race___1'] == "1":
        race.append('race_1')
    elif demographics.loc[i,'race___2'] == "1":
        race.append('race_2')
    elif demographics.loc[i,'race___3'] == "1":
        race.append('race_3')
    elif demographics.loc[i,'race___4'] == "1":
        race.append('race_4')
    elif demographics.loc[i,'race___5'] == "1":
        race.append('race_5')
    elif demographics.loc[i,'race___6'] == "1":
        race.append('race_6')
    else:
        race.append('unknown')
demographics.insert(1,'race',race)
```
